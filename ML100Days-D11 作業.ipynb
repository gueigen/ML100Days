```python
import pandas as pd 
q_df = pd.DataFrame([['male', 'teacher'], ['male', 'engineer'], ['female', None], ['female', 'engineer']],columns=['Sex','Profession'])
q_df
```

```python
q_df1 = q_df.fillna('other')
q_df1
```

```python
from sklearn.preprocessing import LabelEncoder
q_df1['Profession_label'] = LabelEncoder().fit_transform(q_df1['Profession'].values)
q_df1
```

```python
pf = pd.get_dummies(q_df1[['Profession']])
q_df1 = pd.concat([q_df1, pf], axis=1)
q_df1
#應用One-hot Encoding方式編碼，因為職業間無順序之關係。
```
