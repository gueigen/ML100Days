```python
import pandas as pd
import numpy as np
```

```python
score_df = pd.DataFrame([[1,56,66,70], 
                         [2,90,45,34], 
                         [3,45,32,55], 
                         [4,70,77,89], 
                         [5,56,80,70], 
                         [6,60,54,55], 
                         [7,45,70,79], 
                         [8,34,77,76], 
                         [9,25,87,60], 
                         [10,88,40,43]],
                        columns=['student_id','math_score','english_score','chinese_score'])
score_df = score_df.set_index('student_id')
score_df
```

```python
score_mean = score_df.mean(axis=1)
score_mean.iloc[6]
```

```python
score_df1 = score_df.apply(lambda x : x**(0.5)*10)
score_df1.iloc[6]
```

```python
score_df1.mean(axis=0)
```

```python
#新題目
#1.
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
```

```python
# - 依照 fruit 欄位做合併
pd.merge(df1, df2, on='fruit')
```

```python
# - 依照 index 索引做合併
df1.join( df2 )
```
