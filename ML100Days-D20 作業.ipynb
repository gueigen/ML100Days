```python
import matplotlib.pyplot as plt
import numpy as np
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x,y)
plt.show() 
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

# atan2()方法返回atan(y / x)，以弧度表示。
p = np.arctan2(Y,X)
plt.scatter(X,Y, s=50, c=p, alpha=1)
#s為大小，alpha為透明度
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
